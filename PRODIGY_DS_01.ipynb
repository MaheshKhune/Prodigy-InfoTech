```python
import pandas as pd
import matplotlib.pyplot as plt

# Define the path to the uploaded CSV file
csv_path = csv_path = r"D:\Internship Tasks\API_SP.POP.TOTL_DS2_en_csv_v2_76253\API_SP.POP.TOTL_DS2_en_csv_v2_76253.csv"

# Load the dataset while skipping metadata rows
df = pd.read_csv(csv_path, skiprows=4)

# Select relevant columns: Country Name and most recent year's population (2023)
df = df[['Country Name', '2023']].dropna()

# Rename columns for clarity
df.columns = ['Country', 'Population_2023']

# Convert population column to numeric values
df['Population_2023'] = pd.to_numeric(df['Population_2023'], errors='coerce')

# Create a histogram to visualize the population distribution
plt.figure(figsize=(12, 6))
plt.hist(df['Population_2023'], bins=20, color='teal', edgecolor='black', alpha=0.7)
plt.xlabel('Population (2023)')
plt.ylabel('Number of Countries')
plt.title('Distribution of Country Populations in 2023')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
